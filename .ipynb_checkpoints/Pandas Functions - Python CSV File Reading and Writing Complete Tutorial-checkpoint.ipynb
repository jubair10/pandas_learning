## Manipulating Data
* we can pass the index as single to manipulate one data and as an array/list to manipulate multiple data
* we can pass the column as single to manipulate one data and as an array/list to manipulate multiple data
* we can pass the manipulated data as single to manipulate one data and as an array/list to manipulate multiple data
```Python
csv_1.loc[[0,1],["Company","Name"]]=["Manipulated Data","Arithmetic Operators.py"]
csv_1
```

````python


csv_1.loc[[0,1],["Company","Name"]]=["Manipulated Data","Arithmetic Operators.py"]
csv_1
````

* you can also access data using loc
